```python
# Step.1-1 ライブラリのインポート
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

# Step.1-2 教材7.4で作成したデータのインポート
df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# Step2 カテゴリー別の要素カウント＆実行結果の棒グラフ化
num = df['カテゴリー'].value_counts()
num.plot(kind='bar')
plt.title("弁当、デザートなどカテゴリーごとの出現頻度")

# Step3 グラフを表示するコード
plt.show()

# Step4 グループごとの集計と基本統計量の算出
df.groupby('商品番号')['カテゴリー'].count()
df.groupby('商品番号')['カテゴリー'].describe()
```
